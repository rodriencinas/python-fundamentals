```python
#making a basico Bokeh graph

#importing Bokeh
from bokeh.plotting import figure
from bokeh.io import output_file, show

#prepare some data
x = [1,2,3,4,5] 
y = [6,7,8,9,10]

#prepare the output file
output_file("Line.html")

#create a figure object
f = figure()

#create line plot
f.line(x,y)

#show graph
show(f) 
```

```python
#making a basico Bokeh graph - (circle)

#importing Bokeh
from bokeh.plotting import figure
from bokeh.io import output_file, show

#prepare some data
x = [1,2,3,4,5] 
y = [6,7,8,9,10]

#prepare the output file
output_file("Line.html")

#create a figure object
f = figure()

#create line plot
f.circle(x,y, radius=0.5)

#show graph
show(f) 
```

```python
#making a basico Bokeh graph - (triangle)

#importing Bokeh
from bokeh.plotting import figure
from bokeh.io import output_file, show

#prepare some data
x = [1,2,3,4,5] 
y = [6,7,8,9,10]

#prepare the output file
output_file("Line.html")

#create a figure object
f = figure()

#create line plot
f.scatter(marker='triangle', x=x, y=y, size=50)

#show graph
show(f) 
```
